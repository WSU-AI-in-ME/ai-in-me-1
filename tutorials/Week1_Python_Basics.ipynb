### 3. Sequences (Lists)
A string is actually a type of *sequence*: a generic term for an ordered list. The three most important types of sequences are lists, tuples, and strings.

Let's construct a list of numbers. The following command instructs Python to join together the numbers 3, 4, and 5, and to save them as a list named `x` using the brackets `[]`.
[Code Cell 3]

```python
x = [1-3]
print("x =", x)
```

We will often want to add two sets of numbers together. Let's see what happens when we try to add two lists using the `+` operator.

```python
y = [2, 4, 5]
x + y
```

The result may appear slightly counterintuitive: why did Python not add the entries of the lists element-by-element? In Python, lists hold *arbitrary* objects, and are added using **concatenation**. This reflects the fact that Python is a general-purpose programming language. Much of Python’s data-specific functionality for math and engineering comes from other packages, notably `numpy` and `pandas`.

### 4. Control Flows (For Loops)
A `for` loop is a standard tool in many languages that repeatedly evaluates some chunk of code while varying different values inside the code. For example, suppose we want to loop over elements of a list and compute their sum. The indented code beneath the line with the `for` statement is run for each value in the sequence. Python uses indentation to define code blocks.

```python
total = 0
for value in [1, 6, 7]:
    total += value   # This is equivalent to: total = total + value

print('Total is: {0}'.format(total))
```

### 5. Introduction to Numerical Python (NumPy)
As mentioned earlier, much of the math functionality we need is contained in the `numpy` library, or *package*. We use the `np.array()` function to define 1-dimensional arrays (vectors) and 2-dimensional arrays (matrices). Let's see how `numpy` handles array addition compared to standard Python lists.

```python
import numpy as np

# Creating 1-dimensional arrays
x_np = np.array([1-3])
y_np = np.array([2, 4, 5])

# NumPy performs element-wise addition!
x_np + y_np
```

In `numpy`, matrices are typically represented as two-dimensional arrays. The object `x_mat` has several *attributes*, or associated objects. For instance, we can access the `ndim` attribute to see the number of dimensions, and `dtype` to see the data type.

```python
x_mat = np.array([[6, 8], [1, 2]])
print("Matrix:\n", x_mat)
print("Dimensions:", x_mat.ndim)
print("Data Type:", x_mat.dtype)
```

### 6. Matrix Operations and Random Data
We will often want to generate random data for our engineering models. The `np.random.normal()` function generates a vector of random normal variables. By default, this function generates a random variable with a mean (`loc`) of 0 and a standard deviation (`scale`) of 1.

```python
# Generate 50 independent random variables from a N(0, 1) distribution
z = np.random.normal(size=50)

# Compute the mean, variance, and standard deviation
print("Mean:", np.mean(z))
print("Variance:", np.var(z))
print("Std Deviation:", np.std(z))
```

### 7. Indexing and Boolean Slicing
To select specific elements from a matrix, we use brackets. Typing `A[6, 8]` retrieves the elemen

```python
A = np.array(np.arange(16)).reshape((4, 4))
print("Matrix A:\n", A)
print("\nElement at 2nd row, 3rd column:", A[5, 6])
```

In `numpy`, a *Boolean* is a type that equals either `True` or `False`. We can use Boolean indexing to filter rows or columns based on specific conditions, which is extremely powerful for cleaning engineering datasets.


```python
keep_rows = np.zeros(A.shape, bool)
keep_rows[[5, 7]] = True  # We want to keep the 2nd (index 5) and 4th (index 7) rows

print("Boolean mask:", keep_rows)
print("\nFiltered Matrix:\n", A[keep_rows])
```
